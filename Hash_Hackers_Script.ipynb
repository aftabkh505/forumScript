### <font face="Rubik" size="4">» </font><font face="Rubik" size="5">Extract + Upload Frames from Video to ImgBB</font>
<font size="3"><i>Create your own ImgBB API Key here: https://api.imgbb.com/</font>

```
ImgBB_API_Key="" #@param {type:"string"}
path = "" #@param {type:"string"}
!mkdir -p "/content/frames/"

!apt-get install ffmpeg

import os
import base64
import requests
from IPython.display import clear_output

os.environ['inputFile'] = path
!rm -f /content/frames/*

def upload_to_ImgBB(img):
  with open(img, "rb") as file:
    url = "https://api.imgbb.com/1/upload"
    payload = {
        "key": ImgBB_API_Key,
        "image": base64.b64encode(file.read()),
    }
    res = requests.post(url, payload).json()
    return res['data']['url']

clear_output()
!ffmpeg -hide_banner -ss 02:30.0 -i "$inputFile" -vframes 1 -q:v 2 -y "/content/frames/frame1.png"
clear_output()
!ffmpeg -hide_banner -ss 05:00.0 -i "$inputFile" -vframes 1 -q:v 2 -y "/content/frames/frame2.png"
clear_output()
!ffmpeg -hide_banner -ss 07:30.0 -i "$inputFile" -vframes 1 -q:v 2 -y "/content/frames/frame3.png"
clear_output()

%cd "/content/frames/"
print("Uploading Images...")
Screenshots=[]
for i in range(1,4):
  Screenshots.append(upload_to_ImgBB("frame{}.png".format(i)))
%cd "/content/"
clear_output()

text_dump = ""
text_dump += "[b][color=#D5A6D5][size=medium]Screenshots:[/size][/color][/b]\n"
text_dump += "[spoiler][align=center]\n"
for i in range(3):
  text_dump += "[img]{}[/img]\n\n".format(Screenshots[i])
text_dump += "[/align][/spoiler]\n"

print(text_dump)
```
